```python
import time
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import pandas as pd
```

```python
# 4131 pageê¹Œì§€ì„ 

#for helper in range(1, 43): #100page ë‹¨ìœ„ -> ì˜ ëŠê¹€
#for helper in range(1, 85): #50page ë‹¨ìœ„ë¡œ
for helper in range(1, 3):  ###test
    driver = webdriver.Chrome("./chromedriver")
    
    ids = []
    titles = []
    paths = []
    contents = []
    dates = []
    nums = []
    url = ""

    i = (50 * (helper - 1)) + 1
    for page in range(i, 50+i):
    #for page in range(i, 1+i): ###test
        if page == 4132: #pageëŠ” 4131 ê¹Œì§€ë°–ì— ì—†ìŒ.
            break
        url = 'https://www1.president.go.kr/petitions/?c=42&only=2&page={}&order=1'.format(page)
        driver.get(url)
        print("current page : ", page)

        # driver.get(url)ë¡œ ì„œë²„ì—ê²Œ get ìš”ì²­ì„ ë³´ëƒˆìœ¼ë‹ˆ responseë¥¼ ê¸°ë‹¤ë ¤ì•¼í•¨. htmlì„ ë°›ëŠ”ë° ê±¸ë¦¬ëŠ” ìµœì†Œì‹œê°„ -> 1.5s
        time.sleep(1.5)
        soup = bs(driver.page_source, 'html.parser') # 3ì´ˆê°€ ëë‚˜ë©´ ìƒˆë¡œìš´ pageì˜ urlë¡œ ì ‘ì†í•˜ê³  íŒŒì„œê°€ íŒŒì‹±í•œë‹¤.

        # í˜„ pageì—ì„œ petition ìš”ì†Œì— ì ‘ê·¼
        lists = soup.find("div", attrs={"class": "ct_list1"})
        ul = lists.find("ul", attrs={"class": "petition_list"})
        a_lists = ul.find_all("a", attrs={"class": "cb relpy_w"})  # í˜„ì¬ í˜ì´ì§€ì˜ "ì „ì²´ëª©ë¡" ë‚´ë¶€ì— ìˆëŠ” ëª¨ë“  a íƒœê·¸ë“¤ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜.
        id_lists = ul.find_all("div", attrs={"class": "bl_no"})
    
        # pathë¥¼ ì¶”ê°€í•˜ê¸° ì „ì— í˜„ì¬ paths ë¦¬ìŠ¤íŠ¸ì— ëª‡ê°œê°€ ë“¤ì–´ìˆëŠ”ì§€ í™•ì¸í•œë‹¤ => 2 pageê°€ ë˜ë©´ paths ë¦¬ìŠ¤íŠ¸ì˜ 7ë²ˆ indexë¶€í„° ì½ì–´ì•¼í•˜ê¸° ë•Œë¬¸
        path_len = len(paths)
        
        for id in id_lists:
            #print(id.text[3:].strip()) id
            ids.append(id.text[3:].strip())
        
        for a in a_lists: # í˜„ì¬ page ë‚´ì—ì„œ ê° petitionì˜ titleê³¼ pathë¥¼ íŒŒì‹±í•˜ì—¬ paths, titles ë°°ì—´ì— ì €ì¥.
            # print(a.attrs["href"]) path ë°°ì—´ì— ë“¤ì–´ê°ˆ ê²½ë¡œ "/petition/12345"
            paths.append(a.attrs["href"])  # ê° ì²­ì›ì˜ pathë¥¼ ì €ì¥.
            # print(a.text[3:].strip()) titles
            titles.append(a.text[3:].strip())  # ê° ì²­ì›ì˜ titleì„ titles ë°°ì—´ì— ì €ì¥.

        new_paths = paths[path_len:] # page ë³„ë¡œ í•´ë‹¹í•˜ëŠ” pathë§Œ ì½ì–´ì•¼í•˜ë¯€ë¡œ => 2 pageê°€ ë˜ë©´ paths ë¦¬ìŠ¤íŠ¸ì˜ 7ë²ˆ indexë¶€í„° ì½ì–´ì•¼í•˜ê¸° ë•Œë¬¸
        print("new_paths : ", new_paths)
        for path in new_paths: # í˜„ì¬ pageì˜ ëª¨ë“  petitionë“¤ì˜ í•´ë‹¹ urlì— ì ‘ì†í•˜ì—¬ contentë¥¼ íŒŒì‹±.
            url = "https://www1.president.go.kr" + path
            driver.get(url)
            soup = bs(driver.page_source, 'html.parser') # ê° urlì— ëŒ€í•˜ì—¬ soup ê°ì²´ë¥¼ ìƒˆë¡œ ìƒì„±í•œë‹¤.

            # contents ì¶”ê°€í•˜ê¸°.
            content = soup.find("div", attrs={"class":"View_write"})
            data = content.text.strip()
            data = data.replace("\n", "") # \n ë¬¸ìë„ ì œê±°
            #print('ë³¸ë¬¸ ë‚´ìš© : ', data)

            # date ì¶”ê°€í•˜ê¸°.
            date_ul = soup.find("ul", attrs={"class": "petitionsView_info_list"})
            li = date_ul.li.next_sibling.next_sibling
            li_data = li.text[4:]

            # num ì¶”ê°€í•˜ê¸°.
            h2 = soup.find("h2", attrs={"class": "petitionsView_count"})
            num = h2.find('span', attrs={"class": "counter"})
            num = num.text

            contents.append(data) # ê³µë°± ì œê±°í•œ dataë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥.
            dates.append(li_data)
            nums.append(num)
    driver.close()
    # ìµœì¢… ê°’
    print("id : ", ids)
    print("titles : ", titles)
    print("nums ë°°ì—´ì˜ ìˆ˜ë¡œ í™•ì¸í•˜ê¸° ", len(nums)) # numì˜ ê°¯ìˆ˜ê°€ 700ê°œ(100í˜ì´ì§€ * 7ê°œ ì²­ì›ê¸€)ì—¬ì•¼ ì•ˆëº´ë†“ê³  í¬ë¡¤ë§ í•œê±°ì„.
    #print("contents : ", contents)
    print("dates : ", dates)
    print("nums : ", nums)

    my_dictionary = {"id": ids, "title": titles, "date": dates, "num": nums, "contents": contents}
    data = pd.DataFrame(my_dictionary)  # ì „ì²´ ë°ì´í„°ë¥¼ ê¸ì€ listì¸ totalì„ dataframeìœ¼ë¡œ ë³€í™˜ì‹œì¼œì£¼ë©´ì„œ ê° columnì˜ ì´ë¦„ì„ ë¶€ì—¬í•´ì¤ë‹ˆë‹¤.

    str_i = str(i)
    str_i2 = str(i+49)
    datatitle = str_i + '_' + str_i2
    # datatitle = input("Please set the title of this excel file : ")  # ê¸ì–´ì˜¨ ë°ì´í„°ë¥¼ ì €ì¥í•  xlsx íŒŒì¼ì˜ ì´ë¦„ì„ ì§€ì •í•´ì¤ë‹ˆë‹¤.
    data.to_excel("crawling_data/" + datatitle + ".xlsx", engine="xlsxwriter")
    #xlsxwriterë¥¼ ì„í¬íŠ¸ì‹œì¼œì„œ engineì„ ì¶”ê°€í•˜ì§€ ì•Šìœ¼ë©´, data.to_excelì—ì„œ IllegalCharacterError ì—ëŸ¬ê°€ ëœ¬ë‹¤.
```
